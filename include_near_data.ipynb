```python
# 파일 경로
data_path = ['AWS/마포', 'AWS/성북', 'AWS/성동', 'AWS/용산']

# 파일 목록 불러오기
file_list = []
for row in range(4):
    file_list.append(os.listdir(data_path[row]))
```

```python
# df 리스트를 만드는 함수

def make_df_list(data_path, file_list, label):
    df_list = []
    col_list = []
    # label값에 따라 원하는 열 선택
    col_list.append(['일시', '기온(°C)', '풍속(m/s)', '습도(%)', '증기압(hPa)',
                     '이슬점온도(°C)', '현지기압(hPa)', '전운량(10분위)', '중하층운량(10분위)', #'운형(운형약어)',
                     '지면온도(°C)', '강수량(mm)'])
    col_list.append(['일시', '강수량(mm)'])
    col_list.append(['일시', '기온(°C)', '습도(%)', '강수량(mm)'])

    for i in file_list:
        file_path = os.path.join(data_path, i)
        df = pd.DataFrame()
        df = pd.read_csv(file_path, usecols = col_list[label], parse_dates = ['일시'], encoding = 'CP949')
        df_list.append(df)

    return df_list
```

```python
# df 리스트
AWS_마포 = make_df_list(data_path[0], file_list[0], 1)
AWS_성북 = make_df_list(data_path[1], file_list[1], 1)
AWS_성동 = make_df_list(data_path[2], file_list[2], 1)
AWS_용산 = make_df_list(data_path[3], file_list[3], 1)

# 결측 날짜 채우기

AWS_마포_fill_missing = fill_missing_index(AWS_마포)
AWS_성북_fill_missing = fill_missing_index(AWS_성북)
AWS_성동_fill_missing = fill_missing_index(AWS_성동)
AWS_용산_fill_missing = fill_missing_index(AWS_용산)
```
