```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler

from pickle import dump
```

```python
raw_train_ASOS = pd.read_csv('make_data/train_ASOS_include_past.csv', index_col = 0, encoding = 'CP949')
raw_test1_ASOS = pd.read_csv('make_data/test1_ASOS_include_past.csv', index_col = 0, encoding = 'CP949')
raw_test2_ASOS = pd.read_csv('make_data/test2_ASOS_include_past.csv', index_col = 0, encoding = 'CP949')
raw_test3_ASOS = pd.read_csv('make_data/test3_ASOS_include_past.csv', index_col = 0, encoding = 'CP949')

display(raw_train_ASOS)
```

```python
raw_train_AWS = pd.read_csv('make_data/train_AWS_include_past.csv', index_col = 0, encoding = 'CP949')
raw_test1_AWS = pd.read_csv('make_data/test1_AWS_include_past.csv', index_col = 0, encoding = 'CP949')
raw_test2_AWS = pd.read_csv('make_data/test2_AWS_include_past.csv', index_col = 0, encoding = 'CP949')
raw_test3_AWS = pd.read_csv('make_data/test3_AWS_include_past.csv', index_col = 0, encoding = 'CP949')

display(raw_train_AWS)
```

```python
raw_train_add_AWS = pd.concat([raw_train_ASOS, raw_train_AWS], axis = 1)
raw_test1_add_AWS = pd.concat([raw_test1_ASOS, raw_test1_AWS], axis = 1)
raw_test2_add_AWS = pd.concat([raw_test2_ASOS, raw_test2_AWS], axis = 1)
raw_test3_add_AWS = pd.concat([raw_test3_ASOS, raw_test3_AWS], axis = 1)

display(raw_train_add_AWS)
```

```python
# target data 추출

train_target_ASOS = raw_train_ASOS['log_percip']
test1_target_ASOS = raw_test1_ASOS['log_percip']
test2_target_ASOS = raw_test2_ASOS['log_percip']
test3_target_ASOS = raw_test3_ASOS['log_percip']

# 어차피 동일
# train_target_add_AWS = raw_train_add_AWS['log_percip']
# test1_target_add_AWS = raw_test1_add_AWS['log_percip']
# test2_target_add_AWS = raw_test2_add_AWS['log_percip']
# test3_target_add_AWS = raw_test3_add_AWS['log_percip']
```

```python
MM_sc1 = MinMaxScaler()
MM_sc2 = MinMaxScaler()

scaled_train_ASOS = MM_sc1.fit_transform(raw_train_ASOS)
scaled_test1_ASOS = MM_sc1.transform(raw_test1_ASOS)
scaled_test2_ASOS = MM_sc1.transform(raw_test2_ASOS)
scaled_test3_ASOS = MM_sc1.transform(raw_test3_ASOS)

scaled_train_add_AWS = MM_sc2.fit_transform(raw_train_add_AWS)
scaled_test1_add_AWS = MM_sc2.fit_transform(raw_test1_add_AWS)
scaled_test2_add_AWS = MM_sc2.fit_transform(raw_test2_add_AWS)
scaled_test3_add_AWS = MM_sc2.fit_transform(raw_test3_add_AWS)
```

```python
np.save('make_data/scaled_train_ASOS', scaled_train_ASOS)
np.save('make_data/scaled_test1_ASOS', scaled_test1_ASOS)
np.save('make_data/scaled_test2_ASOS', scaled_test2_ASOS)
np.save('make_data/scaled_test3_ASOS', scaled_test3_ASOS)

np.save('make_data/scaled_train_add_AWS', scaled_train_add_AWS)
np.save('make_data/scaled_test1_add_AWS', scaled_test1_add_AWS)
np.save('make_data/scaled_test2_add_AWS', scaled_test2_add_AWS)
np.save('make_data/scaled_test3_add_AWS', scaled_test3_add_AWS)

dump(MM_sc1, open('make_data/MM_sc_ASOS.pkl', 'wb'))
dump(MM_sc2, open('make_data/MM_sc_add_AWS.pkl', 'wb'))
```

```python
np.save('make_data/train_target', train_target_ASOS)
np.save('make_data/test1_target', test1_target_ASOS)
np.save('make_data/test2_target', test2_target_ASOS)
np.save('make_data/test3_target', test3_target_ASOS)
```
