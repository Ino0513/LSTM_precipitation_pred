```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import datetime as dt

from datetime import timedelta, datetime
```

```python
pred_test_target1_ASOS = np.load('modeling_data/ASOS_pred_test1.npy')
pred_test_target2_ASOS = np.load('modeling_data/ASOS_pred_test2.npy')
pred_test_target3_ASOS = np.load('modeling_data/ASOS_pred_test3.npy')

pred_test_target1_AWS = np.load('modeling_data/AWS_pred_test1.npy')
pred_test_target2_AWS = np.load('modeling_data/AWS_pred_test2.npy')
pred_test_target3_AWS = np.load('modeling_data/AWS_pred_test3.npy')

pred_test_target1_P = np.load('modeling_data/P_pred_test1.npy')
pred_test_target2_P = np.load('modeling_data/P_pred_test2.npy')
pred_test_target3_P = np.load('modeling_data/P_pred_test3.npy')

obsv_test_target1 = np.load('modeling_data/obsv_test_target1_P.npy')
obsv_test_target2 = np.load('modeling_data/obsv_test_target2_P.npy')
obsv_test_target3 = np.load('modeling_data/obsv_test_target3_P.npy')
```

```python
obsv_test_target1 = np.insert(obsv_test_target1, 0, 0)
obsv_test_target2 = np.insert(obsv_test_target2, 0, 0)
obsv_test_target3 = np.insert(obsv_test_target3, 0, 0)

print(len(obsv_test_target1))

obsv_test_target1 = np.delete(obsv_test_target1, 4368)
obsv_test_target2 = np.delete(obsv_test_target2, 4368)
obsv_test_target3 = np.delete(obsv_test_target3, 4368)

print(len(obsv_test_target1))
```

```python
# AWS graph

index_test = np.arange(4368)

# graph of test1

b1 = range(0, 4368)
plt.figure(figsize = (11, 4))
plt.plot(index_test[b1], np.power(math.exp(1), pred_test_target1_AWS[b1]) - 1, 'r', label = 'predict')
plt.plot(index_test[b1], np.power(math.exp(1), obsv_test_target1[b1]) - 1, 'b', label = 'observed')

plt.legend(loc='upper right')
plt.grid()
plt.title('Test set1 - 2020')
plt.xlabel('Time (hour)\n')
plt.ylabel('Precipitation (mm)')
plt.ylim([0, 40])
plt.show()

# graph of test2

b2 = range(0, 4368)
plt.figure(figsize = (11, 4))
plt.plot(index_test[b2], np.power(math.exp(1), pred_test_target2_AWS[b2]) - 1, 'r', label = 'predict')
plt.plot(index_test[b2], np.power(math.exp(1), obsv_test_target2[b2]) - 1, 'b', label = 'observed')

plt.legend(loc='upper right')
plt.grid()
plt.title('Test set2 - 2021')
plt.xlabel('Time (hour)\n')
plt.ylabel('Precipitation (mm)')
plt.ylim([0, 80])
plt.show()

# graph of test3

b3 = range(0, 4368)
plt.figure(figsize = (11, 4))
plt.plot(index_test[b3], np.power(math.exp(1), pred_test_target3_AWS[b3]) - 1, 'r', label = 'predict')
plt.plot(index_test[b3], np.power(math.exp(1), obsv_test_target3[b3]) - 1, 'b', label = 'observed')

plt.legend(loc='upper right')
plt.grid()
plt.title('5/1 ~ 10/31')
plt.xlabel('Time (hour)')
plt.ylabel('Precipitation (mm)')
plt.ylim([0, 50])
plt.show()
```
