```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ASOS = pd.read_csv('make_data/ASOS_include_past.csv', encoding = 'CP949')
AWS = pd.read_csv('make_data/AWS_include_past.csv', encoding = 'CP949')

#ASOS.drop(['일시'], axis = 1, inplace = True)
#AWS.drop(['일시'], axis = 1, inplace = True)
```

```python
display(ASOS[:5])
display(AWS[:5])
```

```python
data = pd.concat([ASOS, AWS], axis = 1)
data.drop(data.columns[0], axis = 1, inplace = True)
data.drop(data.columns[17], axis = 1, inplace = True)
```

```python
print(data[0:5])
```

```python
corr = data.corr()
sns.clustermap(corr,
               annot = True,
               cmap = 'RdYlBu_r',
               vmin = -1, vmax = 1)
```

```python
fig, ax = plt.subplots(figsize = (8,8))

mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr,
            cmap = 'RdYlBu_r',
            annot = True,
            mask = mask,
            linewidths = 5,
            cbar_kws = {'shrink' : 0.5},
            vmin = -1, vmax = 1)

plt.show()
```

```python
corr.to_csv('make_data/corr.csv', encoding = 'CP949')
```
